### Filme und Kategorien

1. Du möchtest gerne einen Film ausleihen, der länger als drei Stunden dauert. Welche Filme stehen dir zur Auswahl?

```python
SELECT title, length FROM film 
WHERE length > 180;
```

2. Wie viele Action-Filme haben wir gelistet? Überleg dir, wie du an die Information Action von categories kommst.

```python
SELECT COUNT(*) AS anzahl_actionfilme
FROM film_category
WHERE category_id = 1;
```

3. Welche Filme sind im Lage, wurden aber nie ausgeliehen?

```python
SELECT f.film_id, f.title, i.inventory_id
FROM film f
INNER JOIN inventory i USING(film_id)
LEFT JOIN rental r USING(inventory_id)
WHERE r.rental_id IS NULL;
```

### Schauspieler

1. Wie viele unterschiedliche Schauspieler*in sind in unserer Videothek gelistet?

```python
SELECT count(actor_id) FROM actor;
```

### Kunden

1. Welche Adressen sind keinem Kunden zugewiesen?

```python
SELECT a.address_id, a.address
FROM customer c
RIGHT JOIN address a USING(address_id)
WHERE c.address_id IS NULL;
```

2. Wir wollen ein Special-Offer für alle Kunden anbieten, die aus Ländern kommen, die entweder mit A anfangen oder mit L aufhören. Welche Länder wären das?

```python
SELECT country FROM country 
WHERE country LIKE 'A%' or country LIKE '%l';
```
